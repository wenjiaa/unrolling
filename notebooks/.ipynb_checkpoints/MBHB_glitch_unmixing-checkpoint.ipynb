```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import scipy
import os
from imp import reload
import torch

PATH ='../../../GitHub/LISA_internal/Softwares/' # Give the right location first
sys.path.insert(1,PATH)

import IAE_CNN_TORCH_v2 as cnn

def normalize(X,opt=1):
    Y = X.T
    if opt == 1:
        Y = Y/np.sum(abs(Y),axis=0)
    if opt == 2:
        Y = Y/np.sqrt(np.sum(Y**2,axis=0))
    if opt == 'inf':
        Y = 0.5*Y/np.max(abs(Y),axis=0)
    if opt == 'glob':
        Y = Y/np.max(np.sum(abs(Y),axis=0)) # global normalisation
    return Y.T

def normalize2(X,opt=1):
    Y = np.swapaxes(X,0,2)
    if opt == 1:
        Y = Y/np.sum(abs(Y),axis=(0,1))
    if opt == 2:
        Y = Y/np.sqrt(np.sum(Y**2,axis=(0,1)))
    if opt == 'inf':
        Y = 0.5*Y/np.max(abs(Y),axis=(0,1))
    if opt == 'glob':
        Y = Y/np.max(np.sum(abs(Y),axis=(0,1))) # global normalisation
    return np.swapaxes(Y,0,2)

from jax.lib import xla_bridge
print("Platform used:",xla_bridge.get_backend().platform)
from os.path import exists
```

### Glitch-like model

```python
def glitch_model(N,tau,t0,dt=5):
    
    # t0 should be positive
    # tau in [1,10] for dt = 5
    
    t = np.linspace(0,N-1,N)*dt
    g = (t-t0)/dt*np.exp(-(t-t0)/(dt*tau))  # starts at 0
    
    return g*(g > 0),t
```

```python
N = 10000
t = 2150
dt = 5

tau = 49*np.random.rand(N)+1
t0 = (t-500)*dt*np.random.rand(N)+50

data = np.zeros((N,2150))

for r in range(N):
    g,_ = glitch_model(t,tau[r],t0[r],dt=dt)
    data[r,:] = g

I = np.where(np.isnan(np.mean(data,axis=1)) == False)
data = data[I]
XvalidG = data.reshape(-1,2150,1)
```

```python
modelG = cnn.load_model('small4unmixing_glitches') # Load the glitch model

out = modelG.fast_interpolation(XvalidG[0:1000])

rec =out["XRec"] # Reconstructed signal
L = out["Lambda"][0].detach().numpy() # The parameters in the latent space

i=90
plt.figure(figsize=(15,10))
plt.plot(XvalidG[i,:,0].T,alpha=0.5,lw=4)
plt.plot(rec[i,:,0].detach().numpy().T)
plt.plot(XvalidG[i,:,0].T-rec[i,:,0].detach().numpy().T)
plt.title('Example of FI reconstruction')

plt.figure(figsize=(15,10))
err = -20*np.log10(np.linalg.norm(XvalidG[0:1000]-rec.detach().numpy(),axis=(1,2))/np.linalg.norm(XvalidG[0:1000],axis=(1,2)))
plt.hist(err,100,alpha=0.3,label='FI')
plt.title('Histogram of NMSE in dB')
```
