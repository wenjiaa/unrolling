```python
model_IAE_list = [
    cnn.load_model('Models/cnn_test')
]
Ytest,Atest,Xtest=generate_dataset_mixture(model_IAE_list, 1000, noise_level=[1e-3,1e-1],sigma_a=0.)
dataloader = DataLoader(TensorDataset(torch.from_numpy(Ytest).float(), torch.from_numpy(Atest).float(), torch.from_numpy(Xtest).float()), batch_size=1000)
nmse_x=[]
err_a=[]

bins_x = np.linspace(0, 150,25)
bins_a = np.linspace(-10, 0,25)

plt.figure(figsize=(7,5))
for fname in ['Models/models0721/model_GLFBS_CNN_L6_Ba133']:
    model = GLPALM.load_model(fname, device=device)
    Apred = []
    Xpred = []
    for data in dataloader:
        Y,A,X = data
        Apred_batch, Lam_pred_batch = model.forward(Y.to(device))
        Apred_batch = Apred_batch[-1].cpu().detach().numpy()
        Xpred_batch = model.psi(Lam_pred_batch[-1]).cpu().detach().numpy()
        Apred.append(Apred_batch)
        Xpred.append(Xpred_batch)
    Apred = np.vstack(Apred)
    Xpred = np.vstack(Xpred)

    nmse_x.append(-20*np.log10(((Xpred-Xtest)**2).sum((2,3))/(Xtest**2).sum((2,3))))
    err_a.append(np.log(np.abs(Apred-Atest).squeeze()+1e-10))
#     err_a = np.log((np.abs(Apred-Atest)/(np.abs(Atest)+1e-10)).squeeze())
    
    print('GLFBS')
    print('médiane de nmse_x: ',np.median(nmse_x[-1],0))
    print('médiane de err_a: ',np.median(err_a[-1],0))

    for j in range(Xtest.shape[1]):
        plt.subplot(1,Xtest.shape[1],j+1)
        plt.hist(nmse_x[-1][:,j],bins_x,alpha=0.4,label='GLFBS')  
        plt.title('$x_%d$'%(j+1))
        plt.legend()
        
print('FI:')
Apred_FI = []
Xpred_FI = []

for j in range(Xtest.shape[1]):
    PhiX,PhiE = model_IAE_list[j].encode(torch.tensor(Ytest).float())
    B,_ = model_IAE_list[j].interpolator(PhiX,PhiE)
    Xpred = model_IAE_list[j].decode(B).detach().numpy()

    for _ in range(0):
        PhiX,PhiE = model_IAE_list[j].encode(torch.tensor(Xpred).float())
        B,_ = model_IAE_list[j].interpolator(PhiX,PhiE)
        Xpred = model_IAE_list[j].decode(B).detach().numpy()

    nmse_x_FI = -20*np.log10(((Xpred-Xtest[:,j])**2).sum((1,2))/(Xtest[:,j]**2).sum((1,2)))

    plt.subplot(1,Xtest.shape[1],j+1)
    plt.hist(nmse_x_FI,bins_x,alpha=0.4,label="FI")  
    plt.title('$x_%d$'%(j+1))
    plt.legend()
    print('médiane de nmse_x: %.2f dB' %(np.median(nmse_x_FI,0)))
# plt.savefig('./Figures/GLPALM_CNN_x_sigma0.1.png', bbox_inches='tight')
```

```python
def BCD_LPALM(set_loader, fnameLPALM, itmax = 1000, eps = 1e-6, device='cpu'):
    X_est_list = []
    A_est_list = []
    itmax_list = []
    
    for data in set_loader: # For all mini-batches
        Y,A,X = data
        Y = Y.to(device)

        X_est_mb = torch.zeros(X.size())
        A_est_mb = torch.zeros(A.size())

        it_max_mb = torch.zeros(Y.size()[0])

        for j in range(Y.size()[0]): # For all samples in the current mini-batch
            X_est = torch.ones(X_est_mb[0].size(), device=device) / 2.
            A_est = torch.ones(A_est_mb[0].size(), device=device) / 3.
            gamma = 1.
            

            X_est_prev = X_est.clone() # For the stopping criterion
            A_est_prev = A_est.clone()

            it = 0
            while(torch.norm(X_est-X_est_prev, p = 'fro')/torch.norm(X_est, p = 'fro') > eps or torch.norm(A_est-A_est_prev, p = 'fro')/torch.norm(A_est, p = 'fro') > eps or it < 2) and it < itmax:
                if it>0:
                    X_est_prev = X_est.clone()
                    A_est_prev = A_est.clone()                    
                model = GLPALM.load_model(fnameLPALM, device=device)
                with torch.no_grad():
                    a, lam = model.forward(Y[j].unsqueeze(0)/gamma) # y/gamma
                X_est = model.psi(lam[-1]) 
                A_est = a[-1]
                AX = torch.einsum('ijl,ijkl->kl', A_est, X_est)
                gamma = torch.einsum('kl,kl->l', Y[j], AX)/torch.einsum('kl,kl->l', AX, AX)
    
                it += 1
#                 print(torch.norm(X_est-X_est_prev, p = 'fro')/torch.norm(X_est, p = 'fro'))

            X_est_mb[j] = X_est.cpu().detach()
            A_est_mb[j] = A_est.cpu().detach()
            it_max_mb[j] = it
            print(it)

        X_est_list.append(X_est_mb)
        A_est_list.append(A_est_mb)
        itmax_list.append(it_max_mb)

    return X_est_list, A_est_list, itmax_list
```

```python
def BCD(set_loader, fnameIAE_list, itmax = 1000, alpha=0.9, eps = 1e-6, device='cpu'):
    X_est_list = []
    A_est_list = []
    itmax_list = []
    
    for data in set_loader: # For all mini-batches
        Y,A,X = data
        Y = Y.to(device)

        X_est_mb = torch.zeros(X.size())
        A_est_mb = torch.zeros(A.size())
        it_max_mb = torch.zeros(Y.size()[0])

        for j in range(Y.size()[0]): # For all samples in the current mini-batch
            X_est = torch.ones(X_est_mb[0].size(), device=device) / 2.
            A_est = torch.ones(A_est_mb[0].size(), device=device) / 3.

            X_est_prev = X_est.clone() # For the stopping criterion
            A_est_prev = A_est.clone()

            it = 0
            while(torch.norm(X_est-X_est_prev, p = 'fro')/torch.norm(X_est, p = 'fro') > eps or torch.norm(A_est-A_est_prev, p = 'fro')/torch.norm(A_est, p = 'fro') > eps or it < 2) and it < itmax:
                print('It ext:',torch.norm(X_est-X_est_prev, p = 'fro')/torch.norm(X_est, p = 'fro'), torch.norm(A_est-A_est_prev, p = 'fro')/torch.norm(A_est, p = 'fro'))
                if it>0:
                    X_est_prev = X_est.clone()
                    A_est_prev = A_est.clone()
                  
                # FBS iterations for each x_i 
                it_fbs_sum = 0
                for i, fname in enumerate(fnameIAE_list): 
                    print(i)
                    iae = cnn.load_model(fname, device=device)
                    lam_est =  torch.ones(iae.num_ap, device=device)/iae.num_ap
                    lam_est_prev = lam_est.clone()
                    it_fbs = 0
                    while(torch.norm(lam_est-lam_est_prev, p = 'fro')/torch.norm(lam_est_prev, p = 'fro') > eps or it_fbs < 2) and it_fbs < itmax:
                        # print(torch.norm(lam_est-lam_est_prev, p = 'fro')/torch.norm(lam_est_prev, p = 'fro'))
                        if it_fbs>0:
                            lam_est_prev = lam_est.clone()
                                            
                        _,PhiE = iae.encode(iae.anchorpoints)
                        B = []
                        lam_copy = lam_est.clone().requires_grad_()
                        for r in range(iae.NLayers):
                            B.append(torch.einsum('k,kjl->jl', lam_copy, PhiE[iae.NLayers-r-1])[None,:])
                        psi_lam = iae.decode(B)
                        gradients =  Y[j]-torch.einsum('jkl,jl->kl', psi_lam, A_est).unsqueeze(0)
                        lam_est = lam_est + torch.autograd.grad(psi_lam, lam_copy, grad_outputs=gradients)[0].detach()

                        it_fbs = it_fbs + 1
                    with torch.no_grad():
                        _,PhiE = iae.encode(iae.anchorpoints)
                        B = []
                        for r in range(iae.NLayers):
                            B.append(torch.einsum('k,kjl->jl', lam_est, PhiE[iae.NLayers-r-1])[None,:])
                        X_est[i] = iae.decode(B)

                    it_fbs_sum += it_fbs
                
                
                A_est = torch.einsum('kl,kjl->jl', Y[j], torch.pinverse(X_est.squeeze()).unsqueeze(-1))
    
                it += 1
            
            X_est_mb[j] = X_est.cpu().detach()
            A_est_mb[j] = A_est.cpu().detach()
            it_max_mb[j] = it

        X_est_list.append(X_est_mb)
        A_est_list.append(A_est_mb)
        itmax_list.append(it_max_mb)

    return X_est_list, A_est_list, itmax_list
```
