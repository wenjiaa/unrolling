```python
import numpy as np
import matplotlib.pyplot as plt

import os
while os.path.split(os.getcwd())[1] != "unrolling":
    os.chdir("..")
    if os.getcwd() == "/":
        raise ValueError()
print("Current working directory: {}".format(os.getcwd()))

import sys
PATH ='./codes/'
sys.path.insert(1,PATH)

import IAE_CNN_TORCH_v2 as cnn
import LPALM_LS
import LPALM
import LFBS
from FBS import FBS
from utils import generate_dataset

import torch
from torch.utils.data import TensorDataset, DataLoader

import pytorch_lightning as pl
import time

from sklearn.decomposition import PCA

import pandas as pd
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

# Fixing random state for reproducibility
np.random.seed(19680801)


if torch.cuda.is_available():
    device = 'cuda'
    kwargs = {}
    acc = "gpu"
    Xpus_per_trial = 1
else:
    device = 'cpu'
    kwargs = {}
    acc = 'cpu'
    Xpus_per_trial = 1

print("device USED: ",device)
```

```python
t = time.time()
fname = './Models/archive_LFBS/model_LFBS_L4_random_init_last'
model_LFBS = LFBS.load_model(fname, device=device)

model_IAE = cnn.load_model(model_LFBS.iae_fname)

plt.figure(figsize=(21,5))
i=1
for _noise_level in [1e-1, 1e-2, 1e-3]:
    Ytest, Atest, Xtest = generate_dataset(model_IAE, 1000, noise_level = [_noise_level, _noise_level], amplitude_range=1.)

    dataloader = DataLoader(TensorDataset(torch.from_numpy(Ytest).float(), torch.from_numpy(Xtest).float()), batch_size=50)
    X_est_list, itmax_list, _ = FBS(dataloader, model_IAE, alpha = .9, itmax = 1000)
    print('max iterations of FBS: ', torch.max(torch.cat(itmax_list)))
    X_pred = torch.cat(X_est_list).detach().numpy()
    X_pred_LFBS = model_LFBS.forward(torch.tensor(Ytest).float().to(device))[-1].cpu().detach().numpy()
    X_ref = model_IAE.fast_interpolation(Ytest)['XRec'].detach().numpy() # only apply IAE
    
    nmse_LFBS = -10*np.log10(np.sum((X_pred_LFBS-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    nmse_FBS = -10*np.log10(np.sum((X_pred-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    nmse_IAE = -10*np.log10(np.sum((X_ref-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    print('LFBS: ', np.median(nmse_LFBS))
    print('FBS: ', np.median(nmse_FBS))
    print('IAE: ', np.median(nmse_IAE))
    bins = np.linspace(np.min([nmse_LFBS, nmse_FBS, nmse_IAE]), np.max([nmse_LFBS, nmse_FBS, nmse_IAE]), 25)
    plt.subplot(1,3,i)
    plt.hist(nmse_LFBS,bins,alpha=0.4,label='LFBS')
    plt.hist(nmse_FBS,bins,alpha=0.4,label='FBS')
    plt.hist(nmse_IAE,bins,alpha=0.4,label='IAE')
    plt.xlabel('NMSE(dB))')
    plt.title('SNR = %.2f dB' % (20*np.log10(0.5/_noise_level)))
    plt.legend()
    i += 1
# plt.savefig('./Figures/hist_LFBS.png', bbox_inches='tight')
plt.show()
    
print('Computation time with gpu: %.3f s'% (time.time()-t))
```

Learning curves showed that 
- NLayers=2 or NLayers=3 is optimal for LPALM model in the cas $a \in [0.5, 2.0]$
- NLayers=2 is optimal for LPALM model in the cas $a=1$

Check it by histograms:

```python
t = time.time()
list_fname=['Models/models0524/model_LPALM_L1_a1',
           'Models/models0524/model_LPALM_L2_a1',
           'Models/models0524/model_LPALM_L3_a1',
           'Models/models0524/model_LPALM_L4_a1',
           'Models/models0524/model_LPALM_L5_a1',
           'Models/models0524/model_LPALM_L6_a1']

for _noise_level in [1e-1, 1e-2, 1e-3]:
    plt.figure(figsize=(12,5))
    for fname in list_fname:
        model_LPALM = LPALM.load_model(fname, device=device)
        model_IAE = cnn.load_model(model_LPALM.iae_fname)

        Ytest, Atest, Xtest = generate_dataset(model_IAE, 1000, noise_level = [_noise_level, _noise_level], amplitude_range=1.)
        dataloader = DataLoader(TensorDataset(torch.from_numpy(Ytest).float(), torch.from_numpy(Xtest).float()), batch_size=50)

        a_LPALM, x_LPALM = model_LPALM.forward(torch.tensor(Ytest).float().to(device))
        a_LPALM = a_LPALM[-1].cpu().detach().numpy()
        x_LPALM = x_LPALM[-1].cpu().detach().numpy()
        nmse_x_LPALM = -10*np.log10(np.sum((x_LPALM-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
        nmse_a_LPALM = np.log10(np.sum((a_LPALM-Atest)**2, 1)/ np.sum(Atest**2, 1)+1e-16)

        plt.subplot(121)
        bins = np.linspace(0,60,25)
        plt.hist(nmse_x_LPALM,bins,alpha=0.4,label=fname[24:])
        plt.xlabel('NMSE(dB)')
        plt.title('SNR = %.2f dB' % (20*np.log10(0.5/_noise_level)))
        plt.legend()

        plt.subplot(122)
        bins = np.linspace(-16,1,25)
        plt.hist(nmse_a_LPALM,bins,alpha=0.4,label=fname[24:])
        plt.xlabel('log(relative error on a)')
        plt.title('LPALM (SNR=%.2f dB)'%(20*np.log10(0.5/_noise_level)))
        plt.legend()

#     plt.savefig('./Figures/hist%d_'%(-np.log10(_noise_level))+'.png', bbox_inches='tight')
            
print('Computation time with gpu: %.3f s'% (time.time()-t))
```

- **NLayer=1** is perfect for amplitudes' estimation in the case $a=1$ since the initialization by ones gives directly the ground-truth $a$. 
- NLayers=2 is optimal for both x-estimation and a-estimation.
