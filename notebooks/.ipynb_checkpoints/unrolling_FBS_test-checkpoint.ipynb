```python
plt.figure(figsize=(12,8))
for n in np.arange(2, 11):
    loss_by_layer = np.zeros(n)
    log_dir = "./CKPT/lightning_logs/NLayers = "+str(n)+"/"
    event_accumulator = EventAccumulator(log_dir)
    event_accumulator.Reload()
    for i in np.arange(n):
        y_key = "validation_layer_loss/layer"+str(i+1)
        loss_by_layer[i] = event_accumulator.Scalars(y_key)[-1].value
    plt.plot(np.arange(n)+1, loss_by_layer, label="NLayers = "+str(n))
plt.legend()
plt.xlabel('Layers')
plt.ylabel('Loss')
# plt.savefig('./Figures/Loss_by_layer_without_control.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,8))
for n in np.arange(4, 11):
    loss_by_layer = np.zeros(n)
    log_dir = "./CKPT/lightning_logs/NLayers = "+str(n)+" (with intermediate control)/"
    event_accumulator = EventAccumulator(log_dir)
    event_accumulator.Reload()
    for i in np.arange(n):
        y_key = "validation_layer_loss/layer"+str(i+1)
        loss_by_layer[i] = event_accumulator.Scalars(y_key)[-1].value
    plt.semilogy(np.arange(n)+1, loss_by_layer, label="NLayers = "+str(n))
plt.legend()
plt.xlabel('Layers')
plt.ylabel('Loss')
plt.savefig('./Figures/Loss_by_layer_with_control_zoom.png', bbox_inches='tight')
plt.show()
```

```python
import time
t = time.time()
fname = './Models/model_LFBS_L6_random_init_last_v2'
model_LFBS = load_model(fname, device=device)

model_IAE = cnn.load_model(model_LFBS.iae_fname)
# _noise_level=1e-1

for _noise_level in [1e-1, 1e-2, 1e-3]:
    Ytest, Atest, Xtest = generate_dataset(model_IAE, 1000, noise_level = [_noise_level, _noise_level])

    dataloader = DataLoader(TensorDataset(torch.from_numpy(Ytest).float(), torch.from_numpy(Xtest).float()), batch_size=50)
    X_est_list, itmax_list, _ = FBS(dataloader, model_IAE, alpha = .9, itmax = 1000)
    print('max iterations of FBS: ', torch.max(torch.cat(itmax_list)))
    X_pred = torch.cat(X_est_list).detach().numpy()
    X_pred_LFBS = model_LFBS.forward(torch.tensor(Ytest).float().to(device))[-1].cpu().detach().numpy()
    X_ref = model_IAE.fast_interpolation(Ytest)['XRec'].detach().numpy() # only apply IAE
    
    nmse_LFBS = -10*np.log10(np.sum((X_pred_LFBS-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    nmse_FBS = -10*np.log10(np.sum((X_pred-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    nmse_IAE = -10*np.log10(np.sum((X_ref-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    
    bins = np.linspace(np.min([nmse_LFBS, nmse_FBS, nmse_IAE]), np.max([nmse_LFBS, nmse_FBS, nmse_IAE]), 25)

    plt.hist(nmse_LFBS,bins,alpha=0.4,label='LFBS')
    plt.hist(nmse_FBS,bins,alpha=0.4,label='FBS')
    plt.hist(nmse_IAE,bins,alpha=0.4,label='IAE')
    plt.xlabel('NMSE(dB))')
    plt.title('SNR = %.2f dB' % (20*np.log10(0.5/_noise_level)))
    plt.legend()
#     plt.savefig('./Figures/hist%d_'%(-np.log10(_noise_level))+fname[20:]+'.png', bbox_inches='tight')
    plt.show()
    
print('Computation time with gpu: %.3f s'% (time.time()-t))
```

```python
# compare two LFBS models
t = time.time()
fname = './Models/model_LFBS_L6_random_init_last_v1'
model_LFBS = load_model(fname, device=device)

model_IAE = cnn.load_model(model_LFBS.iae_fname)

fname = './Models/model_LFBS_L4_random_init_last'
model_LFBS2 = load_model(fname, device=device)

# _noise_level=1e-1

for _noise_level in [1e-1, 1e-2, 1e-3]:
    Ytest, Atest, Xtest = generate_dataset(model_IAE, 1000, noise_level = [_noise_level, _noise_level])

    dataloader = DataLoader(TensorDataset(torch.from_numpy(Ytest).float(), torch.from_numpy(Xtest).float()), batch_size=50)
    X_pred_LFBS = model_LFBS.forward(torch.tensor(Ytest).float().to(device))[-1].cpu().detach().numpy()
    X_pred_LFBS2 = model_LFBS2.forward(torch.tensor(Ytest).float().to(device))[-1].cpu().detach().numpy()
    
    nmse_LFBS = -10*np.log10(np.sum((X_pred_LFBS-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    nmse_LFBS2 = -10*np.log10(np.sum((X_pred_LFBS2-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
    bins = np.linspace(np.min([nmse_LFBS, nmse_LFBS2]), np.max([nmse_LFBS, nmse_LFBS2]), 25)

    plt.hist(nmse_LFBS,bins,alpha=0.4,label='model1')
    plt.hist(nmse_LFBS2,bins,alpha=0.4,label='model2')
    plt.xlabel('NMSE(dB))')
    plt.title('SNR = %.2f dB' % (20*np.log10(0.5/_noise_level)))
    plt.legend()
#     plt.savefig('./Figures/hist%d_'%(-np.log10(_noise_level))+fname[20:]+'.png', bbox_inches='tight')
    plt.show()
    
print('Computation time with gpu: %.3f s'% (time.time()-t))
```

```python
# compare LFBS models
t = time.time()
list_fname = ['Models/model_LFBS_L5_random_init_last',
              'Models/model_LFBS_L6_random_init_last_v2']

# _noise_level=1e-1
for _noise_level in [1e-1, 1e-2, 1e-3]:
    list_nmse=[]
    for fname in list_fname:
        model_LFBS = load_model(fname, device=device)

        model_IAE = cnn.load_model(model_LFBS.iae_fname)

        Ytest, _, Xtest = generate_dataset(model_IAE, 1000, noise_level = [_noise_level, _noise_level])

        dataloader = DataLoader(TensorDataset(torch.from_numpy(Ytest).float(), torch.from_numpy(Xtest).float()), batch_size=50)
        X_pred_LFBS = model_LFBS.forward(torch.tensor(Ytest).float().to(device))[-1].cpu().detach().numpy()

        nmse_LFBS = -10*np.log10(np.sum((X_pred_LFBS-Xtest)**2, (1,2))/ np.sum(Xtest**2, (1,2)))
        list_nmse.append(nmse_LFBS)

    plt.figure()
    bins = np.linspace(np.array(list_nmse).min(),np.array(list_nmse).max(),25)
    for i in range(len(list_nmse)):
        plt.hist(list_nmse[i],bins,alpha=0.4,label=list_fname[i][18:])
    plt.xlabel('NMSE(dB))')
    plt.title('SNR = %.2f dB' % (20*np.log10(0.5/_noise_level)))
    plt.legend()
#     plt.savefig('./Figures/hist%d_'%(-np.log10(_noise_level))+'non_neg.png', bbox_inches='tight')
            
print('Computation time with gpu: %.3f s'% (time.time()-t))
```

```python
model_LFBS = load_model('./Models/model_LFBS_L5')
_noise_level = 1e-1
Ytest, _, Xtest = generate_dataset(cnn.load_model(model_LFBS.iae_fname), 1, noise_level = [_noise_level, _noise_level])
Xtest_projected = pca.transform(Xtest[:,:,0])

x_LFBS = model_LFBS.forward(torch.from_numpy(Ytest).float())
x_LFBS = torch.vstack(x_LFBS).squeeze().detach().numpy()
x_LFBS_projected = pca.transform(x_LFBS)
for i in range(model_LFBS.NLayers):
    plt.scatter(x_LFBS_projected[i,0], x_LFBS_projected[i,1],color='orange', marker='o', alpha=(i+1)/model_LFBS.NLayers)
plt.scatter(x_LFBS_projected[i,0], x_LFBS_projected[i,1],color='orange', marker='o', alpha=(i+1)/model_LFBS.NLayers, label="LFBS with last-layer loss")
print(np.linalg.norm(x_LFBS[-1]-Xtest.squeeze())/np.linalg.norm(Xtest.squeeze()))
print(np.linalg.norm(x_LFBS_projected[-1]-Xtest_projected.squeeze())/np.linalg.norm(Xtest_projected.squeeze()))


model_LFBS = load_model('./Models/model_LFBS_L5layer_control')
x_LFBS = model_LFBS.forward(torch.from_numpy(Ytest).float())
x_LFBS = torch.vstack(x_LFBS).squeeze().detach().numpy()
x_LFBS_projected = pca.transform(x_LFBS)
for i in range(model_LFBS.NLayers):
    plt.scatter(x_LFBS_projected[i,0], x_LFBS_projected[i,1],color='g', marker='^', alpha=(i+1)/model_LFBS.NLayers)
plt.scatter(x_LFBS_projected[i,0], x_LFBS_projected[i,1],color='g', marker='^', alpha=(i+1)/model_LFBS.NLayers, label="LFBS with all-layer loss")

dataloader = DataLoader(TensorDataset(torch.from_numpy(Ytest).float(), torch.from_numpy(Xtest).float()), batch_size=32)
_, nb_it, x_FBS = FBS(dataloader, model_IAE, alpha = .9, itmax = 1000, mode='trajectory')
x_FBS = torch.vstack(x_FBS).squeeze().detach().numpy()
x_FBS_projected = pca.transform(x_FBS)
for i in range(x_FBS_projected.shape[0]):
    plt.scatter(x_FBS_projected[i,0], x_FBS_projected[i,1],color='gray', marker='s', alpha=(i+1)/x_FBS_projected.shape[0])
plt.scatter(x_FBS_projected[i,0], x_FBS_projected[i,1],color='gray', marker='s', alpha=(i+1)/x_FBS_projected.shape[0], label="FBS")

print(np.linalg.norm(x_LFBS[-1]-Xtest.squeeze())/np.linalg.norm(Xtest.squeeze()))
print(np.linalg.norm(x_LFBS_projected[-1]-Xtest_projected.squeeze())/np.linalg.norm(Xtest_projected.squeeze()))
print(np.linalg.norm(x_FBS[-1]-Xtest.squeeze())/np.linalg.norm(Xtest.squeeze()))
print(np.linalg.norm(x_FBS_projected[-1]-Xtest_projected.squeeze())/np.linalg.norm(Xtest_projected.squeeze()))

plt.scatter(Xtest_projected[:,0], Xtest_projected[:,1],color='r', marker='*',label="target")
plt.legend() 
plt.savefig("Figures/traj3.png")
plt.show()
```
